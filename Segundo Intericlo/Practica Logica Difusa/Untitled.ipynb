```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

    Creación nuevos antecedentes que contienen variables y miembros del universo.
    
    Creación de las funciones

```python
# New Antecedent/Consequent objects hold universe variables and membership
# functions
temperatura = ctrl.Antecedent(np.arange(0, 61, 2), 'temperatura')
humedad = ctrl.Antecedent(np.arange(10, 71, 2), 'humedad')
RPMotor = ctrl.Consequent(np.arange(0, 61, 2), 'RPMotor')
```

    Funcion de membresía personalizada de entrada para construir interactivamente con un
    antecedente para la temperatura:
    
 

```python
# Custom membership functions can be built interactively with a familiar,
# Pythonic API

temperatura['bajo'] = fuzz.trapmf(temperatura.universe, [-1, 0, 10, 20])
temperatura['medio'] = fuzz.trapmf(temperatura.universe, [10, 20, 40, 50])
temperatura['alto'] = fuzz.trapmf(temperatura.universe, [40, 50, 60, 61])
temperatura.view()
```

    Funcion de membresía personalizada de entrada para construir interactivamente con un
    antecedente para la humedad:
    
   

```python

humedad['bajo'] = fuzz.trapmf(humedad.universe, [-1,10, 20, 30])
humedad['medio'] = fuzz.trapmf(humedad.universe, [20, 30, 50, 60])
humedad['alto'] = fuzz.trapmf(humedad.universe, [50, 60, 70, 71])
humedad.view()
```

    Funcion de membresía personalizada de entrada para construir interactivamente con un
    antecedente para el RPM del Motor:



    

```python
RPMotor['bajo'] = fuzz.trapmf(RPMotor.universe, [-1,0, 10, 20])
RPMotor['medio'] = fuzz.trapmf(RPMotor.universe, [10, 20, 40, 50])
RPMotor['alto'] = fuzz.trapmf(RPMotor.universe, [40, 50, 60, 61])
RPMotor.view()
```

    Creacion de las reglas:


```python
rule1 = ctrl.Rule(temperatura['bajo'] | humedad['alto'], RPMotor['bajo'])
rule2 = ctrl.Rule(temperatura['medio'] | humedad['alto'], RPMotor['medio'])
rule3 = ctrl.Rule(temperatura['alto'] | humedad['alto'], RPMotor['medio'])

rule4 = ctrl.Rule(temperatura['bajo'] | humedad['medio'], RPMotor['bajo'])
rule5 = ctrl.Rule(temperatura['medio'] | humedad['medio'], RPMotor['bajo'])
rule6 = ctrl.Rule(temperatura['alto'] | humedad['medio'], RPMotor['medio'])

rule7 = ctrl.Rule(temperatura['bajo'] | humedad['bajo'], RPMotor['bajo'])
rule8 = ctrl.Rule(temperatura['medio'] | humedad['bajo'],RPMotor['bajo'])
rule9 = ctrl.Rule(temperatura['alto'] | humedad['bajo'], RPMotor['alto'])

```

```python
control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
controlSystemSimulation = ctrl.ControlSystemSimulation(control)
```

```python
controlSystemSimulation.input['temperatura'] = 18
controlSystemSimulation.input['humedad'] = 22
controlSystemSimulation.compute()
s=float(controlSystemSimulation.output['RPMotor'])
print(s)
#RPMotor.view(sim=tipping)
plt.show(RPMotor.view(sim=controlSystemSimulation))
```

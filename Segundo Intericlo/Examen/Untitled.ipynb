### Agregar similitudes de coseno al gráfico
    Crear una relación [: SIMILARIDAD] entre cada persona en el gráfico, donde su similitud de coseno es 
    una propiedad de la relación. La consulta que logra esto es:

```python
graph.run("MATCH (p1:Person)-[x:CLASIFICADO]->(m:Movie)<-[y:CLASIFICADO]-(p2:Person)"+" "
              +"WITH  SUM(x.rating * y.rating) AS xyDotProduct,"+" "
              +"SQRT(REDUCE(xDot = 0.0, a IN COLLECT(x.rating) | xDot + a^2)) AS xLength,"+" "
              +"SQRT(REDUCE(yDot = 0.0, b IN COLLECT(y.rating) | yDot + b^2)) AS yLength, p1, p2"+" "
              +"MERGE (p1)-[s:SIMILARITY]-(p2) SET   s.similarity = xyDotProduct / (xLength * yLength)").data()
```
